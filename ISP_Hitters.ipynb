```python
import pandas as pd
# Defining file paths
hh_path = 'raw-datasets/hof_hitters.xlsx'
nhh_path = 'raw-datasets/non-hof_hitters.xlsx'

# Reading files into dataframes
hof_hitters = pd.read_excel(hh_path)
non_hof_hitters = pd.read_excel(nhh_path)
```

```python
# Hitting data
hof_hitters_stats = hof_hitters[['Player','From','To','G','PA','H','1B','2B','3B','HR','RBI','BA','OBP','SLG','OPS','OPS+','TB']]
non_hof_hitters_stats = non_hof_hitters[['Player','From','To','G','PA','H','1B','2B','3B','HR','RBI','BA','OBP','SLG','OPS','OPS+','TB']]

# Printing results
print(hof_hitters_stats, non_hof_hitters_stats)
```

```python
# # Setting criteria for to rule out pitchers who were generally not good at hitting
# # NOTE: The dataset for Non-HOF hitters all have minimum of GP <= 162.
p_hof_hitters = hof_hitters_stats[hof_hitters_stats['BA'] >= 0.240]
p_non_hof_hitters = non_hof_hitters_stats[non_hof_hitters_stats['G'] >= 50] # Redundant, just to create new dataframe for consistancy in naming scheme.

# Printing results
print(p_hof_hitters)
print(p_non_hof_hitters)
```

```python
# Getting mean, max, and min values of each column
avg_hof_hitters = p_hof_hitters.mean(numeric_only=True).round(3)
min_hof_hitters = p_hof_hitters.min(numeric_only=True).round(3)
max_hof_hitters = p_hof_hitters.max(numeric_only=True).round(3)
median_hof_hitters = p_hof_hitters.median(numeric_only=True).round(3)
q25_hof_hitters = p_hof_hitters.quantile(0.25, numeric_only=True).round(3)
q75_hof_hitters = p_hof_hitters.quantile(0.75, numeric_only=True).round(3)

avg_non_hof_hitters = p_non_hof_hitters.mean(numeric_only=True).round(3)
min_non_hof_hitters = p_non_hof_hitters.min(numeric_only=True).round(3)
max_non_hof_hitters = p_non_hof_hitters.max(numeric_only=True).round(3)
median_non_hof_hitters = p_non_hof_hitters.median(numeric_only=True).round(3)
q25_non_hof_hitters = p_hof_hitters.quantile(0.25, numeric_only=True).round(3)
q75_non_hof_hitters = p_hof_hitters.quantile(0.75, numeric_only=True).round(3)

# Printing results
print("HoF Hitters - Averages:\n", avg_hof_hitters)
print("HoF Hitters - Minimums:\n", min_hof_hitters)
print("HoF Hitters - Maximums:\n", max_hof_hitters)
print("HoF Hitters - Medians:\n", median_hof_hitters)
print("HoF Hitters - 25th Percentile:\n", q25_hof_hitters)
print("HoF Hitters - 75th Percentile:\n", q75_hof_hitters)

print("Non-HoF Hitters - Averages:\n", avg_non_hof_hitters)
print("Non-HoF Hitters - Minimums:\n", min_non_hof_hitters)
print("Non-HoF Hitters - Maximums:\n", max_non_hof_hitters)
print("Non-HoF Hitters - Medians:\n", median_non_hof_hitters)
print("Non-HoF Hitters - 25th Percentile:\n", q25_non_hof_hitters)
print("Non-HoF Hitters - 75th Percentile:\n", q75_non_hof_hitters)
```

#  HOF Baseline

## Minimums
* **H >= 111**
* 1B >= 83
* 2B >= 16
* 3B >= 5
* HR >= 2
* RBI >= 65
* **BA >= 0.248**
* **OBP >= 0.262**
* **SLG >= 0.306**
* **OPS >= 0.582**
* **OPS+ >= 58**
* TB >= 158

## Medians
* **H >= 2304**
* 1B >= 1529
* 2B >= 391
* 3B >= 77
* HR >= 148
* RBI >= 1138
* **BA >= 0.305**
* **OBP >= 0.377**
* **SLG >= 0.464**
* **OPS >= 0.840**
* **OPS+ >= 129**
* TB >= 3528

## 25th Percentile
* **H >= 1778**
* 1B >= 1219
* 2B >= 290
* 3B >= 50
* HR >= 65
* RBI >= 839
* **BA >= 0.285**
* **OBP >= 0.377**
* **SLG >= 0.429**
* **OPS >= 0.797**
* **OPS+ >= 115**
* TB >= 2636

```python
# Examining certain stats

p_hof_hitters_ba = p_hof_hitters[['BA']]
p_non_hof_hitters_ba = p_non_hof_hitters[['BA']]

p_hof_hitters_obp = p_hof_hitters[['OBP']]
p_non_hof_hitters_obp = p_non_hof_hitters[['OBP']]

p_hof_hitters_slg = p_hof_hitters[['SLG']]
p_non_hof_hitters_slg = p_non_hof_hitters[['SLG']]

p_hof_hitters_ops = p_hof_hitters[['OPS']]
p_non_hof_hitters_ops = p_non_hof_hitters[['OPS']]

p_hof_hitters_opsplus = p_hof_hitters[['OPS+']]
p_non_hof_hitters_opsplus = p_non_hof_hitters[['OPS+']]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(5,3, figsize=(18,25))


# BA
# BA for HOF Hitters
axs[0, 0].hist(p_hof_hitters_ba, bins=10, color='red')
axs[0, 0].set_xlabel('BA')
axs[0, 0].set_ylabel('Number of Players')
axs[0, 0].set_title('Batting Average for HOF Hitters (Higher is better)')
axs[0, 0].set_xlim(0,0.5)

# BA for Non-HOF Hitters
axs[0, 1].hist(p_non_hof_hitters_ba, bins=50, color='cyan')
axs[0, 1].set_xlabel('BA')
axs[0, 1].set_ylabel('Number of Players')
axs[0, 1].set_title('Batting Average for Non-HOF Hitters (Higher is better)')
axs[0, 1].set_xlim(0,0.5)

# BA for both overlayed
axs[0, 2].hist(p_hof_hitters_ba, bins=10, color='red', density=True)
axs[0, 2].hist(p_non_hof_hitters_ba, bins=50, color='cyan', alpha=0.6, density=True)
axs[0, 2].set_xlabel('BA')
axs[0, 2].set_ylabel('Proportion of Players from Group')
axs[0, 2].set_title('Batting Average for HOF and Non-HOF Hitters (Higher is better)')
axs[0, 2].set_xlim(0,0.5)


# OBP
# OBP for HOF Hitters
axs[1, 0].hist(p_hof_hitters_obp, bins=15, color='lime')
axs[1, 0].set_xlabel('OBP')
axs[1, 0].set_ylabel('Number of Players')
axs[1, 0].set_title('OBP for HOF Hitters (Higher is better)')
axs[1, 0].set_xlim(0,0.6)

# OBP for Non-HOF Hitters
axs[1, 1].hist(p_non_hof_hitters_obp, bins=30, color='magenta')
axs[1, 1].set_xlabel('OBP')
axs[1, 1].set_ylabel('Number of Players')
axs[1, 1].set_title('OBP for Non-HOF Hitters (Higher is better)')
axs[1, 1].set_xlim(0,0.6)

# OBP for both overlayed
axs[1, 2].hist(p_hof_hitters_obp, bins=15, color='lime', density=True)
axs[1, 2].hist(p_non_hof_hitters_obp, bins=30, color='magenta', alpha=0.6, density=True)
axs[1, 2].set_xlabel('OBP')
axs[1, 2].set_ylabel('Proportion of Players from Group')
axs[1, 2].set_title('OBP for HOF and Non-HOF Hitters (Higher is better)')
axs[1, 2].set_xlim(0,0.6)


# SLG
# SLG for HOF Hitters
axs[2, 0].hist(p_hof_hitters_slg, bins=15, color='blue')
axs[2, 0].set_xlabel('SLG')
axs[2, 0].set_ylabel('Number of Players')
axs[2, 0].set_title('SLG for HOF Hitters (Higher is better)')
axs[2, 0].set_xlim(0,0.75)

# SLG for Non-HOF Hitters
axs[2, 1].hist(p_non_hof_hitters_slg, bins=50, color='gold')
axs[2, 1].set_xlabel('SLG')
axs[2, 1].set_ylabel('Number of Players')
axs[2, 1].set_title('SLG for Non-HOF Hitters (Higher is better)')
axs[2, 1].set_xlim(0,0.75)

# SLG for both overlayed
axs[2, 2].hist(p_hof_hitters_slg, bins=15, color='blue', density=True)
axs[2, 2].hist(p_non_hof_hitters_slg, bins=50, color='gold', alpha=0.6, density=True)
axs[2, 2].set_xlabel('SLG')
axs[2, 2].set_ylabel('Proportion of Players from Group')
axs[2, 2].set_title('SLG for HOF and Non-HOF Hitters (Higher is better)')
axs[2, 2].set_xlim(0,0.75)


# OPS
# OPS for HOF Hitters
axs[3, 0].hist(p_hof_hitters_ops, bins=15, color='sienna')
axs[3, 0].set_xlabel('OPS')
axs[3, 0].set_ylabel('Number of Players')
axs[3, 0].set_title('OPS for HOF Hitters (Higher is better)')
axs[3, 0].set_xlim(0,1.2)

# OPS for Non-HOF Hitters
axs[3, 1].hist(p_non_hof_hitters_ops, bins=50, color='greenyellow')
axs[3, 1].set_xlabel('OPS')
axs[3, 1].set_ylabel('Number of Players')
axs[3, 1].set_title('OPS for Non-HOF Hitters (Higher is better)')
axs[3, 1].set_xlim(0,1.2)

# OPS for both overlayed
axs[3, 2].hist(p_hof_hitters_ops, bins=15, color='sienna', density=True)
axs[3, 2].hist(p_non_hof_hitters_ops, bins=50, color='greenyellow', alpha=0.6, density=True)
axs[3, 2].set_xlabel('OPS')
axs[3, 2].set_ylabel('Proportion of Players from Group')
axs[3, 2].set_title('OPS for HOF and Non-HOF Hitters (Higher is better)')
axs[3, 2].set_xlim(0,1.2)


# OPS+
# OPS+ for HOF Hitters
axs[4, 0].hist(p_hof_hitters_opsplus, bins=15, color='darkorchid')
axs[4, 0].set_xlabel('OPS+')
axs[4, 0].set_ylabel('Number of Players')
axs[4, 0].set_title('OPS+ for HOF Hitters (Higher is better)')
axs[4, 0].set_xlim(0,230)

# OPS+ for Non-HOF Hitters
axs[4, 1].hist(p_non_hof_hitters_opsplus, bins=50, color='dimgrey')
axs[4, 1].set_xlabel('OPS+')
axs[4, 1].set_ylabel('Number of Players')
axs[4, 1].set_title('OPS+ for Non-HOF Hitters (Higher is better)')
axs[4, 1].set_xlim(0,230)

# OPS+ for both overlayed
axs[4, 2].hist(p_hof_hitters_opsplus, bins=15, color='darkorchid', density=True)
axs[4, 2].hist(p_non_hof_hitters_opsplus, bins=50, color='dimgrey', alpha=0.6, density=True)
axs[4, 2].set_xlabel('OPS+')
axs[4, 2].set_ylabel('Proportion of Players from Group')
axs[4, 2].set_title('OPS+ for HOF and Non-HOF Hitters (Higher is better)')
axs[4, 2].set_xlim(0,230)


# Displaying the data
plt.tight_layout()
plt.show
```
