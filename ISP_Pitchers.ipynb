```python
import pandas as pd
# Defining file paths
hp_path = 'raw-datasets/hof_pitchers.xlsx'
nhp_path = 'raw-datasets/non-hof_pitchers.xlsx'

# Reading files into dataframes
hof_pitchers = pd.read_excel(hp_path)
non_hof_pitchers = pd.read_excel(nhp_path)
```

```python
# Pitching data
hof_pitchers_stats = hof_pitchers[['Player','From','To','G','ERA','IP','ERA+','FIP','WHIP','H9','BB9','SO9','SO/BB']]
non_hof_pitchers_stats = non_hof_pitchers[['Player','From','To','G','ERA','IP','ERA+','FIP','WHIP','H9','BB9','SO9','SO/BB']]

# Printing results
print(hof_pitchers_stats, non_hof_pitchers_stats)
```

```python
# Setting criteria for pitchers only (50+ G)
# NOTE: The dataset for Non-HOF Pitchers all have minimum of GP <= 162.
# NOTE: HOF Pitchers are filtered by 50+ G to remove position players who have pitched a few innings over their careers but are not pitchers.
p_hof_pitchers = hof_pitchers_stats[hof_pitchers_stats['G'] >= 50]
p_non_hof_pitchers = non_hof_pitchers_stats[non_hof_pitchers_stats['G'] >= 50] # Redundant, just to create new dataframe for consistancy in naming scheme.

# Printing results
print(p_hof_pitchers)
print(p_non_hof_pitchers)
```

```python
# Getting mean, max, and min values of each column
avg_hof_pitchers = p_hof_pitchers.mean(numeric_only=True).round(3)
min_hof_pitchers = p_hof_pitchers.min(numeric_only=True).round(3)
max_hof_pitchers = p_hof_pitchers.max(numeric_only=True).round(3)
median_hof_pitchers = p_hof_pitchers.median(numeric_only=True).round(3)
q25_hof_pitchers = p_hof_pitchers.quantile(0.25, numeric_only=True).round(3)
q75_hof_pitchers = p_hof_pitchers.quantile(0.75, numeric_only=True).round(3)

avg_non_hof_pitchers = p_non_hof_pitchers.mean(numeric_only=True).round(3)
min_non_hof_pitchers = p_non_hof_pitchers.min(numeric_only=True).round(3)
max_non_hof_pitchers = p_non_hof_pitchers.max(numeric_only=True).round(3)
median_non_hof_pitchers = p_non_hof_pitchers.median(numeric_only=True).round(3)
q25_non_hof_pitchers = p_hof_pitchers.quantile(0.25, numeric_only=True).round(3)
q75_non_hof_pitchers = p_hof_pitchers.quantile(0.75, numeric_only=True).round(3)

# Printing results
print("HoF Pitchers - Averages:\n", avg_hof_pitchers)
print("HoF Pitchers - Minimums:\n", min_hof_pitchers)
print("HoF Pitchers - Maximums:\n", max_hof_pitchers)
print("HoF Pitchers - Medians:\n", median_hof_pitchers)
print("HoF Pitchers - 25th Percentile:\n", q25_hof_pitchers)
print("HoF Pitchers - 75th Percentile:\n", q75_hof_pitchers)

print("Non-HoF Pitchers - Averages:\n", avg_non_hof_pitchers)
print("Non-HoF Pitchers - Minimums:\n", min_non_hof_pitchers)
print("Non-HoF Pitchers - Maximums:\n", max_non_hof_pitchers)
print("Non-HoF Pitchers - Medians:\n", median_non_hof_pitchers)
print("Non-HoF Pitchers - 25th Percentile:\n", q25_non_hof_pitchers)
print("Non-HoF Pitchers - 75th Percentile:\n", q75_non_hof_pitchers)
```

#  HOF Baseline

## Minimums and Maximums
* **ERA <= 3.9**
* IP >= 320
* **ERA+ >= 103**
* **FIP <= 4.35**
* **WHIP <= 1.555**
* H9 <= 10.5
* BB9 <= 4.7
* SO9 >= 2.3
* **SO/BB >= 0.770**

## Medians
* **ERA <= 3.06**
* IP >= 3507
* **ERA+ >= 125**
* **FIP <= 3.13**
* **WHIP <= 1.188**
* H9 <= 8.20
* BB9 <= 2.40
* SO9 >= 5.30
* **SO/BB >= 2.110**

## 25th Percentile
* **ERA <= 2.72**
* IP >= 2398
* **ERA+ >= 115**
* **FIP <= 2.815**
* **WHIP <= 1.137**
* H9 <= 7.75
* BB9 <= 2.05
* SO9 >= 3.90
* **SO/BB >= 1.675**

```python
# Examining certain stats

p_hof_pitchers_eraplus = p_hof_pitchers[['ERA+']]
p_non_hof_pitchers_eraplus = p_non_hof_pitchers[['ERA+']]

p_hof_pitchers_fip = p_hof_pitchers[['FIP']]
p_non_hof_pitchers_fip = p_non_hof_pitchers[['FIP']]

p_hof_pitchers_whip = p_hof_pitchers[['WHIP']]
p_non_hof_pitchers_whip = p_non_hof_pitchers[['WHIP']]

p_hof_pitchers_sobb = p_hof_pitchers[['SO/BB']]
p_non_hof_pitchers_sobb = p_non_hof_pitchers[['SO/BB']]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4,3, figsize=(18,20))


# ERA+
# ERA+ for HOF Pitchers
axs[0, 0].hist(p_hof_pitchers_eraplus, bins=20, color='red')
axs[0, 0].set_xlabel('ERA+')
axs[0, 0].set_ylabel('Number of Players')
axs[0, 0].set_title('ERA+ for HOF Pitchers (Higher is better)')
axs[0, 0].set_xlim(70,210)

# ERA+ for Non-HOF Pitchers
axs[0, 1].hist(p_non_hof_pitchers_eraplus, bins=30, color='cyan')
axs[0, 1].set_xlabel('ERA+')
axs[0, 1].set_ylabel('Number of Players')
axs[0, 1].set_title('ERA+ for Non-HOF Pitchers (Higher is better)')
axs[0, 1].set_xlim(70,210)

# ERA+ for both overlayed
axs[0, 2].hist(p_hof_pitchers_eraplus, bins=20, color='red', density=True)
axs[0, 2].hist(p_non_hof_pitchers_eraplus, bins=30, color='cyan', alpha=0.6, density=True)
axs[0, 2].set_xlabel('ERA+')
axs[0, 2].set_ylabel('Proportion of Players from Group')
axs[0, 2].set_title('ERA+ for HOF and Non-HOF Pitchers (Higher is better)')
axs[0, 2].set_xlim(70,210)


# FIP
# FIP for HOF Pitchers
axs[1, 0].hist(p_hof_pitchers_fip, bins=20, color='lime')
axs[1, 0].set_xlabel('FIP')
axs[1, 0].set_ylabel('Number of Players')
axs[1, 0].set_title('FIP for HOF Pitchers (Lower is better)')
axs[1, 0].set_xlim(1,6)

# FIP for Non-HOF Pitchers
axs[1, 1].hist(p_non_hof_pitchers_fip, bins=30, color='magenta')
axs[1, 1].set_xlabel('FIP')
axs[1, 1].set_ylabel('Number of Players')
axs[1, 1].set_title('FIP for Non-HOF Pitchers (Lower is better)')
axs[1, 1].set_xlim(1,6)

# FIP for both overlayed
axs[1, 2].hist(p_hof_pitchers_fip, bins=20, color='lime', density=True)
axs[1, 2].hist(p_non_hof_pitchers_fip, bins=30, color='magenta', alpha=0.6, density=True)
axs[1, 2].set_xlabel('FIP')
axs[1, 2].set_ylabel('Proportion of Players from Group')
axs[1, 2].set_title('FIP for HOF and Non-HOF Pitchers (Lower is better)')
axs[1, 2].set_xlim(1,6)


# WHIP
# WHIP for HOF Pitchers
axs[2, 0].hist(p_hof_pitchers_whip, bins=20, color='blue')
axs[2, 0].set_xlabel('WHIP')
axs[2, 0].set_ylabel('Number of Players')
axs[2, 0].set_title('WHIP for HOF Pitchers (Lower is better)')
axs[2, 0].set_xlim(0.8,2.0)

# WHIP for Non-HOF Pitchers
axs[2, 1].hist(p_non_hof_pitchers_whip, bins=30, color='gold')
axs[2, 1].set_xlabel('WHIP')
axs[2, 1].set_ylabel('Number of Players')
axs[2, 1].set_title('WHIP for Non-HOF Pitchers (Lower is better)')
axs[2, 1].set_xlim(0.8,2.0)

# WHIP for both overlayed
axs[2, 2].hist(p_hof_pitchers_whip, bins=20, color='blue', density=True)
axs[2, 2].hist(p_non_hof_pitchers_whip, bins=30, color='gold', alpha=0.6, density=True)
axs[2, 2].set_xlabel('WHIP')
axs[2, 2].set_ylabel('Proportion of Players from Group')
axs[2, 2].set_title('WHIP for HOF and Non-HOF Pitchers (Lower is better)')
axs[2, 2].set_xlim(0.8,2.0)


# SO/BB
# SO/BB for HOF Pitchers
axs[3, 0].hist(p_hof_pitchers_sobb, bins=16, color='sienna')
axs[3, 0].set_xlabel('SO/BB')
axs[3, 0].set_ylabel('Number of Players')
axs[3, 0].set_title('SO/BB for HOF Pitchers (Higher is better)')
axs[3, 0].set_xlim(0.4,7.6)

# SO/BB for Non-HOF Pitchers
axs[3, 1].hist(p_non_hof_pitchers_sobb, bins=30, color='greenyellow')
axs[3, 1].set_xlabel('SO/BB')
axs[3, 1].set_ylabel('Number of Players')
axs[3, 1].set_title('SO/BB for Non-HOF Pitchers (Higher is better)')
axs[3, 1].set_xlim(0.4,7.6)

# SO/BB for both overlayed
axs[3, 2].hist(p_hof_pitchers_sobb, bins=16, color='sienna', density=True)
axs[3, 2].hist(p_non_hof_pitchers_sobb, bins=30, color='greenyellow', alpha=0.6, density=True)
axs[3, 2].set_xlabel('SO/BB')
axs[3, 2].set_ylabel('Proportion of Players from Group')
axs[3, 2].set_title('SO/BB for HOF and Non-HOF Pitchers (Higher is better)')
axs[3, 2].set_xlim(0.4,7.6)


# Displaying the data
plt.tight_layout()
plt.show
```
